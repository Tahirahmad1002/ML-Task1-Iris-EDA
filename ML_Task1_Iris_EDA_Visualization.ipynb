# Project: Machine Learning Internship - Task 1
 Objective: Load, inspect, and visualize the Iris dataset to understand data trends and distributions.

# 1. Data Loading and Initial Inspection
We use Pandas to load the dataset and verify its structure. This helps us ensure there are no missing values before we start our analysis

# 2. Data Visualization
We use Seaborn to create scatter plots and box plots. This allows us to see how different flower species are grouped and to identify any outliers in the data.

```python
# --- STEP 1: Import the tools ---
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- STEP 2: Load the data ---
# This pulls the data from the internet and puts it into a table called 'df'
df = sns.load_dataset('iris')

# --- STEP 3: Inspect the data ---
print("--- DATA PREVIEW ---")
print(df.head())  # Shows the first 5 rows

print("\n--- DATA SHAPE ---")
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.")

print("\n--- SUMMARY STATISTICS ---")
print(df.describe()) # Shows mean, min, max, etc.


```

```python
# --- STEP 4: Visualizations ---

# 1. Scatter Plot Matrix (Relationships)
# This shows how every measurement relates to the others, colored by species
sns.pairplot(df, hue='species', height=2.5)
plt.suptitle("Relationship between Features", y=1.02)
plt.show()
```

```python
# 2. Box Plot (Finding Outliers)
# This helps us see if any specific flower measurements are "weirdly" high or low
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Checking for Outliers in Iris Features")
plt.show()
```

```python
# 3. Histogram (Distribution)
# Let's look specifically at Petal Length distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['petal_length'], bins=20, kde=True, color='purple')
plt.title("Distribution of Petal Lengths")
plt.xlabel("Length (cm)")
plt.ylabel("Frequency")
plt.show()
```
